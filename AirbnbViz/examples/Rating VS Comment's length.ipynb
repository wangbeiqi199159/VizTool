```python
import pandas as pd
import numpy as np
#from textblob import TextBlob
import matplotlib.pyplot as plt
%matplotlib inline
from math import pi
from bokeh.plotting import figure, output_notebook, show
from bokeh.charts import Bar, output_file, show
output_notebook() 
```

```python
df = pd.read_csv('comments_rating.csv', sep='\t',index_col = False)
del df['Unnamed: 0']
df.head()
```

```python
df['Comments'] = df['Comments'].astype(str)
```

```python
comments = df['Comments']
```

```python
comment_lengths = []
for i in range(0, len(comments)):
    print ("this is the", + i,"th sentence")
    comment_lengths.append(len(comments[i]))
```

```python
df['comment_length'] = comment_lengths
```

```python
df.head()
```

```python
df_by_rating = df.groupby(['Rating']).mean()
df_by_rating.head()
```

```python
from bokeh.plotting import figure, show, output_file

output_file('rating_vs_comment_length.html')

p = figure(plot_width=800, plot_height=500, title = "Guests' Rating Score VS. Length of Comments")

p.line(df_by_rating.index, df_by_rating.comment_length, color="firebrick",alpha=0.6, line_width=4)
show(p)
```
